```python
import json
import csv
```

```python
data = []
for line in open('internet-archive1.wane', 'r'):
    data.append(json.loads(line))
```

```python
data[1]
```

```python
len(data)
```

```python
import networkx as nx
G=nx.Graph()
```

```python
for item in data:
    G.add_node(item['url'], bipartite=0)
```

top_nodes are the urls

```python
G.number_of_nodes()
```

```python
for item in data:
    G.add_nodes_from(item['named_entities']['persons'], bipartite=1)
```

```python
edges = []
for item in data:
    for index in range(len(item['named_entities']['persons'])):
        l = [[item['url'], item['named_entities']['persons'][index]]]
        edges.extend(tuple(l))
```

```python
G.number_of_nodes()
```

```python
G.add_edges_from(edges)
```

```python
G.number_of_edges()
```

```python
nx.write_graphml(G,"./graph2.graphml")
```

```python
from networkx.algorithms import bipartite
```

```python
nx.is_connected(G)
```

```python
bottom_nodes, top_nodes = bipartite.sets(G)
```

bottom_nodes are the people, top nodes are the urls

```python
top_nodes = set(n for n,d in G.nodes(data=True) if d['bipartite']==0)
```

```python
bottom_nodes = set(G) - top_nodes
```
